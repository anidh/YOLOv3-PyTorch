```python
import json
import cv2
import os
json_file = open("bbox-annotations.json")
```

```python
annotations = json.load(json_file)
```

```python
gt_path = "gt_annotations"
for j in range(0, len(annotations['images'])):
    img_id = annotations['images'][j]['id']
    img = cv2.imread("images\\"+annotations['images'][j]['file_name'])
    a = []
    for i in range(0,len(annotations['annotations'])):
        if annotations['annotations'][i]["image_id"] == img_id:
            x1 = annotations['annotations'][i]["bbox"][0]
            y1 = annotations['annotations'][i]["bbox"][1]
            x2 = annotations['annotations'][i]["bbox"][2] + annotations['annotations'][i]["bbox"][0] 
            y2 = annotations['annotations'][i]["bbox"][3] + annotations['annotations'][i]["bbox"][1]
            cv2.rectangle(img, (x1,y1),(x2,y2),(0,0,255),2)
            
            if annotations['annotations'][i]["category_id"] == 1:
                temp = ["person",x1,y1,x2,y2]
            if annotations['annotations'][i]["category_id"] == 2:
                temp = ["car",x1,y1,x2,y2]
                
            a.append(temp)
                
    mf = open(os.path.join(gt_path,annotations['images'][j]['file_name'][:-4]) + '.txt', 'w')
    text = ''
    for i in range(0,len(a)): 
        append_string = str(a[i][0]) + ' ' + str(a[i][1])+ ' '+str(a[i][2])+ ' '+str(a[i][3])+ ' '+str(a[i][4])+'\n'
        text += append_string
    mf.write(text)
    mf.close()
            
    cv2.imshow("a", img)
    cv2.waitKey(1)
```
